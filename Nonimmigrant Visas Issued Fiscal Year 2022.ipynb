```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Upload CSV
df = pd.read_csv('Downloads/Nonimmigrant Visas Issued Fiscal Year 2022.csv')

# Show first 10 lines.
df.head(10)
```

```python
# Remove null values.
df = df.dropna()

# Show first 10 lines.
df.head(10)
```

```python
# Creating a new CSV
new_doc = "Documents/Python/Nonimmigrant_Visas_Cleaned_rev2.csv"
df.to_csv (new_doc, index=False)
```

```python
# Getting the data types of columns

data_type = df.dtypes
    
print("Column data types:")
print(data_type)
```

```python
# Generating the descriptive summary of the data
df.describe()
```

```python
# Filtering data for the B-2 visa.

b2 = df[df['B-2']>0]

# Sorting in ascending orde.
top10_b2 = b2.sort_values(by='B-2', ascending=False).head(10)

# Results
print("Countries that applied most for type B-2 visas")
print(top10_b2[['Nationality', 'B-2']])

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(top10_b2['Nationality'], top10_b2 ['B-2'], color='purple')
plt.xlabel('Nationality')
plt.ylabel('Quantity')
plt.title ('Countries that applied most for type B-2 visas')
plt.xticks(rotation = 45, ha='right')
plt.tight_layout()

plt.show()
```

```python
# Filtering data for the F visa.
b2 = df[df['F']>0]

# Sorting in ascending order.
top10_b2 = b2.sort_values(by='F', ascending=False).head(10)

# Results
print("Countries that applied most for type F visas")
print(top10_b2[['Nationality', 'F']])

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(top10_b2['Nationality'], top10_b2 ['F'], color='orange')
plt.xlabel('Nationality')
plt.ylabel('Quantity')
plt.title ('Countries that applied most for type F visas')
plt.xticks(rotation = 45, ha='right')
plt.tight_layout()

plt.show()
```
